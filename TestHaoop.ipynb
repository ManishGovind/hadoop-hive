```python
# Replace 'ghcnd-stations.txt' and 'states.txt' with the actual paths to your files
input_file_path = 'ghcnd-stations.txt'
states_file_path = 'ghcnd-states.txt'
output_file_path = 'output.csv'

# Create a dictionary to map state names to abbreviations
state_mapping = {}

# Read the abbreviations and state names from the states.txt file
with open(states_file_path, 'r') as states_file:
    for line in states_file:
        
        abbreviation, state_name = line.split(maxsplit=1)
        print(state_name)
        state_mapping[abbreviation] = state_name.strip()
       

# Open the input file and create the output file
with open(input_file_path, 'r') as input_file, open(output_file_path, 'w') as output_file:
    # Write header to the output CSV file
    output_file.write('ID,NAME,STATE\n')

    # Process line by line
    for line in input_file:
        # Split the line into individual values
        values = line.split()
      
        # Check if there are enough values in the line
        if len(values) >= 9 :
            # Extract 'ID', 'NAME', and 'STATE' values
            station_id = values[0]
            station_name = values[5]  # Join NAME column values
            state = values[4]

            # Replace state name with abbreviation if available
            state_abbreviation = state_mapping.get(state, state)

            # Write the extracted information to the output CSV file
            if station_id[:2] in ['MX','CA','US']:
                
                output_file.write(f"{station_id},{station_name},{state_abbreviation}\n")
                print(f"{station_id},{station_name},{state_abbreviation}")
```
